# 연속확률변수

# 1.연속형 균일 분포

<font size="4">

균일 분포는 이산형인 경우와 연속형인 경우로 나눌 수 있다. 이산형 균일 분포와 연속형 균일 분포의 차이는 확률 변수 $X$가 이산형 확률 변수인지 혹은 연속형 확률 변수인지의 차이이다.

<font size="4">

확률 변수 $X$가 **연속형 균일 분포(continuous uniform distribution)** 를 따를 때 확률 밀도 함수는 다음과 같다.

$$
X\sim U(a,b)
$$
$$
f_X(x)=\frac{1}{b-a}, \quad x \in [a,b]
$$

연속형 균일 분포의 확률 밀도 함수는 위와 같으며 위 식에서 확률 변수 $X$는 $a$와 $b$ 사이의 값을 가지며 $f_X(x)$는 $\frac{1}{b-a}$로 동일한 것을 알 수 있다.

<font size="4">

연속형 균일 분포의 기댓값과 분산은 다음과 같다.
$$
E(X)=\frac{b+a}{2}
$$
$$
Var(X)=\frac{(b-a)^2}{12}
$$

# 2.정규분포

<font size="4">

정규분포는 연속형 확률 분포에서 가장 기본이 되는 분포이다. 확률 변수 $X$가 **정규 분포(normal distribution)** 를 따를 때, 평균이 $\mu$, 분산이 $\sigma^2$인 정규 분포의 확률 밀도 함수는 다음과 같다.
$$
X \sim N(\mu,\sigma^2)
$$
$$
f_X(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-{\frac{1}{2}}(\frac{x-\mu}{\sigma})^2},\quad -\infty<x<\infty
$$

<font size="4">

정규 분포의 평균과 분산은 다음과 같다.

$$
E(X)=\mu
$$
$$
Var(X)=\sigma^2
$$
정규 분포에서 평균 $\mu$는 분포의 위치를 정하는 파라미터로 location parameter라고 부르고, 분산 $\sigma^2$은 분포의 평균으로부터의 흩어짐 정도를 나타내는 파라미터로 scale parameter라고 부른다.

분산이 작은 경우 평균을 중심으로 모여있고, 분산이 큰 경우 값이 넓게 분포되어있다.

## 표준 정규 분포

<font size="4">

정규 분포에서 평균이 $0$, 분산이 $1$인 경우, **표준 정규 분포(standard normal distribution)** 를 따른다고 하며 다음과 같은 확률 밀도 함수를 따른다.

$$
X \sim N(0,1)
$$
$$
f_X(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}x^2}, \quad -\infty<x<\infty
$$


## 정규 근사

<font size="4">

주어진 데이터가 정규 분포를 따르지 않더라도, 데이터의 개수 $n$이 커지면 정규 분포에 가까워지는 현상이 발생할 수 있는데 이를 **정규 근사(normal approximation)** 라고 부른다.

### 이항 분포의 정규 근사

<font size="4">

이항 분포에서 $n$이 커지면 정규 분포에 근접하게 된다. 즉, 확률 변수 $X$가 $Binomial(n,p)$를 따를 때 $n$이 커지면 평균이 $np$이고 분산이 $np(1-p)$인 정규 분포를 따른다. 이를 식으로 나타내면 다음과 같다.
$$
X \sim Binomial(n,p)
$$
$$
X \sim N(np,np(1-p)) \quad as \space n \to \infty
$$


### 포아송 분포의 정규 근사

<font size="4">

이항 분포의 정규 근사와 비슷하게 포아송 분포에도 $\lambda$가 커지면 정규 분포에 근접하게 된다. 확률 변수 $X$가 $Poisson(\lambda)$를 따를 때 $\lambda$가 커지면 평균이 $\lambda$이고 분산이 $\lambda$인 정규분포를 따른다. 이를 식으로 나타내면 다음과 같다.
$$
X \sim Poisson(\lambda)
$$
$$
X \sim N(\lambda,\lambda) \quad as \space \lambda \to \infty
$$


# 3.감마 분포

## 감마 함수

<font size="4">

감마 함수를 이해하기 위해서는 팩터리얼의 개념을 먼저 알아야 한다.
$$
f(x)=x!=x\times (x-1) \times (x-2) \times ... \times 1
$$
본래 팩토리얼 하뭇는 음이 아닌 정수 값만을 가질 수 있는 불연속 함수이다. 따라서 위와 같은 팩토리얼 함수를 사용할 경우 정수가 아닌 실수 값에 대해서는 적용할 수 없다는 한계가 존재한다. 반면 **감마 함수(gamma function)** 는 팩토리얼 함수를 확장시킨 함수로써 팩토리얼 함수를 연속 함수화한 것이라고 생각하면 된다. 감마 함수는 다음과 같이 정의한다.
$$
\Gamma(\alpha)=\int^\infty _0 x^{\alpha-1}e^{-x}dx, \quad \alpha >0
$$

<font size="4">

감마 함수는 다음과 같은 성질을 따른다.

1. $\Gamma(\alpha)=(\alpha-1)!$
2. $\Gamma(\alpha+1)=\alpha \Gamma(\alpha)$
3. $\Gamma(\frac{1}{2})=\sqrt{\pi}$
4. $\Gamma(1)=1$

## 감마 분포

<font size="4">

**감마 분포(gamma distribution)** 를 따르는 확률 변수 $X$는 $\alpha$번째 성공할 때까지의 대기 시간을 의미한다. 확률 변수 $X$가 감마 분포를 따를 때 확률 밀도 함수는 다음과 같다.
$$
X \sim Gamma(\alpha, \beta)
$$
$$
f_X(x)=\frac{1}{\Gamma(\alpha)\beta^{\alpha}}x^{\alpha-1}e^{-\frac{1}{\beta}x}, \quad 0<x<\infty, \quad \alpha>0, \quad \beta>0
$$

위 식에서 확률 변수 $X$의 범위는 $0$부터 무한대까지라는 것을 알 수 있다. 이는 확률 변수 $X$가 성공할 때까지의 대기 시간을 의미한다는 사실을 생각해 보면 대기 시간이 음수일 수는 없으므로 자연스러운 것을 알 수 있다.

<font size="4">

감마 분포의 기댓값과 분산은 다음과 같다.
$$
E(X)=\alpha \beta
$$
$$
Var(X)=\alpha \beta^2
$$

## 감마 분포의 적률 생성 함수

<font size="4">

감마 분포의 적률 생성 함수는 다음과 같다.
$$
M_X(t)=(1-\beta t)^{-\alpha}
$$

<font size="4">

감마 분포를 따르는 확률 변수 $X$를 이용해 감마 분포의 형태를 확인해보자.

$$
f_X(x)=\frac{1}{\Gamma(\alpha)\beta^{\alpha}}x^{\alpha-1}e^{-\frac{1}{\beta}x}, \quad 0<x<\infty, \quad \alpha>0, \quad \beta>0
$$

위와 같은 확률 밀도 함수를 따르는 감마 분포를 가정하자.

```python
def seq(start, stop, step):
    """
    수열
    입력 : start(시작 값), stop(끝 값), step(한 스텝당 증가)
    출력 : res(리스트)
    """
    res = []
    current = start
    while current < stop:
        res.append(current)
        current += step
    return res
```

<font size="4">

$$
 x^{\alpha-1}e^{-x}
$$

```python
def gamma_function(alpha,x):
    """
    감마 함수
    입력 : alpha, x
    출력 : res(실수)
    """

    e = 2.7182818284
    res = (x**(alpha-1))*(e**(-x))
    return res
```

<font size="4">

$$
\Gamma(\alpha)=\int^\infty _0 x^{\alpha-1}e^{-x}dx
$$

```python
def gamma(alpha):
    """
    감마
    입력 : alpha
    출력 : res(실수)
    """

    a = 0
    b = 100
    x_list = seq(0.0001, 100, 0.001)
    gamma_list = []
    for x in x_list:
        y = gamma_function(alpha ,x)
        gamma_list.append(y)
    res = ((b-a)/len(x_list))*sum(gamma_list)
    return res
```

<font size="4">


$$
f_X(x)=\frac{1}{\Gamma(\alpha)\beta^{\alpha}}x^{\alpha-1}e^{-\frac{1}{\beta}x}, \quad 0<x<\infty, \quad \alpha>0, \quad \beta>0
$$

```python
alpha = 2
beta = 1

e = 2.7182818284

x_list = seq(0, alpha*beta*5, 0.01)
p_list = []

for x in x_list :
    prob = (1/(gamma(alpha)*(beta**alpha)))*(x**(alpha-1))*(e**((-1/beta)*x))
    p_list.append(prob)
```

```python
import matplotlib.pyplot as plt

plt.plot(x_list, p_list)
plt.show()
```

# 4.지수분포

<font size="4">

지수 분포는 감마 분포에서 알파가 1인 특수한 경우를 말한다. 확률 변수 $X$가 첫 번째 성공할 때까지의 대기 시간이라고 했을 때, 확률 변수 $X$는 **지수 분포(exponential distribution)** 를 따르며 다음과 같은 확률 밀도 함수를 가진다.
$$
X \sim Exponential(\beta)
$$
$$
f_X(x)=\frac{1}{\beta}e^{-\frac{1}{\beta}x}, \quad x>0
$$

위 식에서 확률 변수 $X$는 $0$보다 크다는 것을 알 수 있다.

<font size="4">

지수 분포의 기댓값과 분산은 다음과 같다.
$$
E(X)=\beta
$$
$$
Var(X)=\beta^2
$$

## 지수 분포의 무기억성

<font size="4">

1. 만약 확률 변수 $X$가 지수 분포를 따른다면, 확률 변수 $X$는 다음과 같은 무기억성을 따른다.

$$
X \sim Exp(\beta)
$$
$$
P(X>s+t|X>t)=P(X>s), \quad for \space all \space s,t>0
$$
$$
P(X>s+t)=P(X>s)P(X>t), \quad for \space all \space s,t>0
$$

2. 만약 확률 변수 $X$가 연속형 확률 변수이고 $X$가 무기억성을 가진다면 확률 변수 $X$는 지수 분포를 따른다.

## 지수 분포의 적률 생성 함수

<font size="4">

$$
M_X(t)=(1-\beta t)^{-1}
$$

<font size="4">

지수 분포를 따르는 확률 변수 $X$를 이용해 지수 분포의 형태를 확인해보자.

$$
f_X(x)=\frac{1}{\beta}e^{-\frac{1}{\beta}x}, \quad x>0
$$

위와 같은 확률 밀도 함수를 따르는 지수 분포를 가정하자.

<font size="4">

$\beta=2$인 지수 분포에 대한 코드

```python
e = 2.7182818284

beta = 2
x_list = seq(0, beta*10, 0.01)
p_list = []

for x in x_list:
    prob = (1/beta)*(e**((-1/beta)*x))
    p_list.append(prob)

plt.plot(x_list, p_list)
plt.show()
```

지수 분포를 따르는 확률 변수 $X$를 이용해 지수 분포의 형태를 확인해보자.

$$

$$

위와 같은 확률 밀도 함수를 따르는 지수 분포를 가정하자.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt

beta = 2
x_list = np.arange(0, beta*10, 0.01)
p_list = []

for x in x_list:
    prob = expon.pdf(x, scale=beta)
    p_list.append(prob)

plt.plot(x_list, p_list)
plt.show()
```

# 5.카이제곱 분포

<font size="4">

$$
X \sim \chi^2(p)
$$
$$
f_X(x)=\frac{1}{\Gamma(\frac{p}{2})2^{\frac{p}{2}}}x^{\frac{p}{2}-1}e^{-\frac{1}{2}x}, \quad 0<x<\infty
$$

<font size="4">

$$
$$

$$
f_X(x)=\frac{1}{\Gamma(\frac{p}{2})2^{\frac{p}{2}}}x^{\frac{p}{2}-1}e^{-\frac{1}{2}x}, \quad 0<x<\infty

# 6.베타분포

### 베타 함수

<font size="4">

**베타 함수(beta function)** 는 이항 계수를 실수 범위로 확장한 것이다. 베타 함수는 다음과 같이 정의한다.
$$
B(\alpha,\beta)=\frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}
$$
$$
B(\alpha,\beta)=\int^1_0x^{\alpha-1}(1-x)^{\beta-1}dx
$$

### 베타 분포

<font size="4">

확률 변수 $X$가 **베타 분포(beta distribution)** 를 따를 때 확률 변수 $X$의 확률 밀도 함수는 다음과 같다.
$$
X\sim Beta(\alpha, \beta)
$$
$$
f_X(x)=\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}, \quad 0<x<1, \quad \alpha>0, \quad \beta>0
$$

지금까지의 확률 분포에서 확률 변수의 범위가 위아래로 제한되어 있는 것은 균일 분포가 있었다. 베타 분포와 균일 분포는 서로 연관되어 있다.


<font size="4">

베타 분포의 기댓값과 분산은 다음과 같다.

$$
E(X)=\frac{\alpha}{\alpha+\beta}
$$
$$
Var(X)=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}
$$

<font size="4">

베타 분포를 따르는 확률 변수 $X$를 이용해 베타 분포의 형태를 확인해보자.
$$
f_X(x)=\frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}x^{\alpha-1}(1-x)^{\beta-1}, \quad 0<x<1, \quad \alpha>0, \quad \beta>0
$$

위와 같은 확률 밀도 함수를 가지는 베타 분포를 가정하자.

<font size="4">

다음 코드는 $\alpha=0.5, \space \beta=0.5$인 베타 분포에 대해 실습하는 코드이다.

```python
alpha = 0.5
beta = 0.5

e = 2.7182818284

x_list = seq(0.01, 1, 0.01)
p_list = []

beta_function = ((gamma(alpha+beta))/(gamma(alpha)*gamma(beta)))
for x in x_list:
    prob = beta_function*(x**(alpha-1))*((1-x)**(beta-1))
    p_list.append(prob)

plt.plot(x_list, p_list)
plt.show()
```

## 베타 분포를 따르는 난수 생성 라이브러리

```python
x_list = np.random.beta(a=0.5, b=0.5, size=10000)
print(x_list)

plt.hist(x_list)
plt.show()
```
